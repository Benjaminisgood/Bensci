## 1. 转化说明
- 默认输入：`data_resourse/assets2/`
- 默认输出：`data_resourse/assets3/`
- 支持输出格式：`json` / `md` / `both`

终端执行：
```bash
python -m bensci.literature_transer --input data_resourse/assets2 --output data_resourse/assets3 --output-format json
```

## 2. OCR 相关参数（可选）
在命令行中可以追加：
- `--ocr-engine auto|tesseract|paddle|easyocr|rapidocr|pypdf2`
- `--ocr-lang eng`
- `--ocr-dpi 300`
- `--ocr-preprocess grayscale|binarize|sharpen`



## 3. 列出可用解析器（可选）


````python
from bensci.transer_tools import describe_parsers

for name, meta in describe_parsers().items():
    print(name, meta)
````


## 4. 执行转化（可选）


````python
from bensci import literature_transer as lt
from bensci import config as cfg

RUN = False  # 改为 True 后执行转化

if RUN:
    lt.convert_path(cfg.XML_SOURCE_DIR, output_dir=cfg.BLOCKS_OUTPUT_DIR, output_format=cfg.TRANSER_OUTPUT_FORMAT)
    print("output:", cfg.BLOCKS_OUTPUT_DIR)
else:
    print("Skip running. Output:", cfg.BLOCKS_OUTPUT_DIR)
````


## 5. 查看输出 JSON（可选）


````python
import json
from pathlib import Path
from bensci import config as cfg

root = Path(cfg.BLOCKS_OUTPUT_DIR)
files = list(root.glob("**/*.json"))
if files:
    sample = files[0]
    data = json.loads(sample.read_text(encoding="utf-8"))
    print("sample:", sample)
    print("keys:", data.keys())
    print("blocks:", len(data.get("blocks", [])))
else:
    print("no json found in:", root)
````

## 输出 JSON 结构说明
- `metadata`：文献元信息（标题、作者、期刊、DOI 等）
- `blocks`：解析后的段落/表格/图像信息

你可以根据 `blocks` 的类型进行二次处理，如筛选 `table` 或 `figure`。
