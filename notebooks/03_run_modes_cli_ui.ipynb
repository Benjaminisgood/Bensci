## 1. 一键运行（run.sh）
项目根目录提供 `run.sh`，支持交互式菜单：
- 1) 打开前端 UI（可视化运行）
- 2) 运行单个任务（交互填写参数）
- 3) 退出项目（停止前端/后端进程）

在终端执行：
```bash
bash run.sh
```

> 注意：run.sh 需要交互输入，在 Jupyter 里运行可能体验不好，建议在终端执行。

## 2. UI 启动与访问
- 选择菜单 `1` 后，默认端口为 `7860`。
- 成功启动后访问：`http://127.0.0.1:7860`（或你设置的端口）。
- UI 日志：`data_resourse/logs/bensci_ui.log`。

## 3. UI 关闭
在 run.sh 菜单中选择 `3` 可停止 UI 进程。



## 4. 查看 UI 日志（可选）


````python
from pathlib import Path

log_path = Path("data_resourse/logs/bensci_ui.log")
if log_path.exists():
    print(log_path.read_text(encoding="utf-8")[-2000:])
else:
    print("UI 日志不存在：", log_path)
````


## 5. 查看 UI 端口与 PID（可选）


````python
from pathlib import Path

pid_file = Path("data_resourse/logs/bensci_ui.pid")
port_file = Path("data_resourse/logs/bensci_ui.port")
print("PID:", pid_file.read_text().strip() if pid_file.exists() else "N/A")
print("PORT:", port_file.read_text().strip() if port_file.exists() else "N/A")
````

## UI 能做什么
- 配置 API Key 与 override
- 以“队列”方式组合多个阶段并一键运行
- 在页面查看统一日志输出

> UI 更适合不熟悉 CLI 的用户；熟悉命令行的用户可以直接执行模块命令。
