## 1. 抽取说明
- 默认输入：`data_resourse/assets3/`
- 默认输出：`data_resourse/assets4/extracted_tap_kinetics_issues.csv`

终端执行：
```bash
python -m bensci.llm_info_extractor
```

你也可以指定模型与输出：
```bash
python -m bensci.llm_info_extractor --model gpt-5-mini --output data_resourse/assets4/custom.csv
```



## 2. 运行抽取（可选）


````python
from bensci import llm_info_extractor as le
from bensci import config as cfg

RUN = False  # 改为 True 后执行

if RUN:
    le.main()
else:
    print("Skip running. Output:", cfg.LLM_EXTRACTION_OUTPUT_PATH)
````


## 3. 自动 schema（可选）
当你需要根据任务动态生成字段时，可启用 `--auto-schema`。
```bash
python -m bensci.llm_info_extractor --auto-schema --schema-sample-size 6 --schema-max-fields 18
```



## 4. 查看输出 CSV（可选）


````python
import csv
from pathlib import Path
from bensci import config as cfg

path = Path(cfg.LLM_EXTRACTION_OUTPUT_PATH)
if path.exists():
    with path.open("r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        rows = [next(reader) for _ in range(5)]
    print("columns:", reader.fieldnames)
    for row in rows:
        print(row.get("article_title"), row.get("doi"))
else:
    print("CSV not found:", path)
````

## 抽取字段与块选择
- 输出字段模板来自 `config.py` 的 `LLM_PROMPTS['extraction']`。
- 系统会用关键词筛选并排序候选 block，再发送给 LLM。
- 如果字段不符合你的任务，可使用 `--auto-schema` 或覆盖模板。
