## 2. 常见问题
- **API 返回 401/403**：检查 `.env` 中的 Key 是否正确。
- **429/Rate Limit**：降低并发、增大 sleep，或减少 `METADATA_MAX_RESULTS`。
- **没有筛选结果**：可能是 query 太严格，或 LLM 返回 `NO`。
- **下载失败**：确认 DOI 是否正确；尝试切换 provider 或 Sci-Hub。
- **解析失败**：检查文件格式是否在解析器支持范围内。
- **OCR 报错**：确认系统安装 `tesseract` / `poppler` 以及 OCR 相关依赖。


## 4. 常用排查动作
- 先检查 `config.override.json` 是否覆盖了路径或 Provider。
- 缩小数据规模：把 `METADATA_MAX_RESULTS` 调到 10/20 进行冒烟测试。
- 使用单个 DOI 测试下载：`python -m bensci.literature_fetcher --doi "10.xxx/xxxx"`
